```python
# Import dependencies
from citipy import citipy
import requests
import json
import pandas as pd
import gmaps

# Google developer API key
from config import g_key

# Access maps with unique API key
gmaps.configure(api_key=g_key)
```

```python
# Import Weather.csv
df = pd.read_csv('output/WeatherPy.csv')
df.head()
```

```python
# creating coordinates and humidity variables for heatmap_layer
coordinates = df[["Lat","Lng"]]
humidity = df["Humidity"]
```

```python
# Create a heat map that displays the humidity for every city from the part I of the homework.

# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px'
}
fig = gmaps.figure(layout=figure_layout)

# Create heatmap
heatmap_layer = gmaps.heatmap_layer(locations = coordinates,weights = humidity)
fig.add_layer(heatmap_layer)
fig
```

```python
# narrowing down DataFrame to find ideal weather condition
ideal_weather_df = df[(df["Max Temp"] > 70) & (df["Max Temp"] < 80) & (df["Wind Speed"] < 10) & (df["Cloudiness"] == 0)]
ideal_weather_df
```

```python
# Use Google Places API to find the first hotel for each city located within 5000 meters of your coordinates.

# creating a list of lat and lng for  request loop
target_coordinates = ideal_weather_df["Lat"].astype(str) + "%2C" + ideal_weather_df["Lng"].astype(str)

# create empty lists for for loop
name = []
city = []
hotel_lat = []
hotel_lng = []

for i in target_coordinates:
    response = requests.get(f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={i}&radius=5000&type=hotel&key={g_key}').json()
    try:
        name.append(json.dumps(response["results"][2]["name"]))
        city.append(json.dumps(response["results"][2]["vicinity"]))
        hotel_lat.append(json.dumps(response["results"][1]["geometry"]["location"]["lat"]))
        hotel_lng.append(json.dumps(response["results"][1]["geometry"]["location"]["lng"]))
    except:
        pass

# create hotel_df for heatmap
hotel_df = pd.DataFrame({
    "name": name,
    "city": city,
    "hotel_lat": hotel_lat,
    "hotel_lng": hotel_lng
})
```

```python
# Plot the hotels on top of the humidity heatmap with each pin containing ***hotel info (can't figure out how to add more than one line)
figure_layout = {
    'width': '800px',
    'height': '600px'
}
fig = gmaps.figure(layout=figure_layout)
heatmap_layer = gmaps.heatmap_layer(locations = coordinates,weights = humidity)
marker_locations = hotel_df[["hotel_lat","hotel_lng"]].astype(float)
marker_box = hotel_df["name"] 
markers = gmaps.marker_layer(marker_locations, hover_text =marker_box, display_info_box=None)
fig.add_layer(heatmap_layer)
fig.add_layer(markers)
fig
```
